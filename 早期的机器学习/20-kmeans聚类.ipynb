```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)   
data=np.random.randint(1,1000,(1000,2))
print(data)

model=KMeans(n_clusters=5)
model.fit(data)

# print(model.cluster_centers_)
# print(model.labels_)

# plt.scatter([1,2,3,4,5,6],[10,20,30,40,50,60],c=[0,1,1,0,1,1])
x=data[:,0]
# print(x)
y=data[:,1]
plt.scatter(x,y,c=model.labels_)
x_center=model.cluster_centers_[:,0]
y_center=model.cluster_centers_[:,1]
plt.scatter(x_center,y_center,marker="o",s=100,c="r")
plt.show()
```
