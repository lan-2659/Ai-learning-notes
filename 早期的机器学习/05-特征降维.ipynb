```python
from sklearn.feature_selection import VarianceThreshold
# 低方差过滤
tool=VarianceThreshold(threshold=1.5)
x=[[10,2],
   [11,6],
   [10,8],
   [10,10],
   [10,19],
   [10,22],
   [10,28]]
x=tool.fit_transform(x)
print(x)
```

```python
# 皮尔逊相关系数
from scipy.stats import pearsonr
x=[10,20,30,40,50]
x2=[10,20,1,30,2]
y=[1,2,3,4,5]
res=pearsonr(x2,y)
print(res.statistic)#相关系数
print(res.pvalue)#p值 越小越好
```

```python
# PCA特征 降维
from sklearn.decomposition import PCA
data = [[21,18,41,51,22,44], 
        [61,32,10,11,31,24], 
        [52,43,14,5,11,14]]
# pca = PCA(n_components=2)#降维到2维 保留2个特征
pca = PCA(n_components=0.8)#信息保留50% 但是不确定会把特征变成几个
data=pca.fit_transform(data)
print(data)
```

```python
# 案例 鸢尾花特征降维
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
x,y=load_iris(return_X_y=True)
pca=PCA(n_components=0.8)
x=pca.fit_transform(x)
print(x.shape)
print(x)
```
